```python
import numpy as np
import matplotlib.pyplot as plt
import scipy 
```

```python
a1=10
a2=3
a3=3
A = 10
```

```python
siglen = 1. #seconds
samplIntrvl = 0.001 #seconds
```

```python
timeVec = np.linspace(0,int(siglen),int(siglen/samplIntrvl))
```

```python
def my_signal(t):
    #t is the time array 

    phaseVec = a1*t + a2*t**2 + a3*t**3
    sigVec = A * np.sin(2*np.pi*phaseVec)
    return sigVec
```

```python
#my_signal(timeVec)
```

```python
plt.plot(timeVec,my_signal(timeVec),'-')
plt.xlabel("$t~(s)$",fontsize=15)
plt.ylabel("Strain",fontsize=15)
#plt.legend()
```
